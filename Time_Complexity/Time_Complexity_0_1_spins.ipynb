```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

bar_values = [1527, 3920, 8746, 6173]
accuracy_values = [0.54, 0.64, 0.57, 0.49]

bar = ax1.bar([1, 2, 3, 4], bar_values, width=0.5, color='lightsteelblue', edgecolor='black', linewidth=1.2)
ax1.set_title('Performance of Deep Learning Models using 0-1 spins data')
ax1.set_xlabel('Deep Learning Models')
ax1.set_xticks([1, 2, 3, 4])
ax1.set_xticklabels(['Bayesian Model', 'Bagging Method', 'Transformers', 'LTSM RNN'])


ax2 = ax1.twinx()
ax2.plot([1, 2, 3, 4], accuracy_values, color='darkslategrey')
ax2.set_ylabel('Accuracy (%)', color="black")
ax1.set_ylabel('Time Complexity of training (in seconds)', color="black")

ax2.tick_params(axis='y', labelcolor='black')
fig.tight_layout()

plt.show()  
```
